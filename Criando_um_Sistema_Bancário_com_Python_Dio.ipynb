```python
menu = """

[1] Depositar
[2] Sacar
[3] Extrato
[4] Sair

=> """

saldo = 0
limite = 500
extrato = ""
numero_saques = 0
LIMITE_SAQUES = 3
emprestimo = 1000
emprestimo_maximo = 1000

while True:

    opcao = input(menu)

    if opcao == "1":
        valor = float(input("Informe o valor do depósito: "))

        if valor > 0:
            saldo += valor


            if emprestimo < emprestimo_maximo:
                valor_reposto = min(valor, emprestimo_maximo - emprestimo)
                emprestimo += valor_reposto
                saldo -= valor_reposto
                extrato += f"Depósito: R$ {valor:.2f} (R$ {valor_reposto:.2f} reposto ao empréstimo)\n"
            else:
                extrato += f"Depósito: R$ {valor:.2f}\n"

            print(f"Depósito realizado com sucesso! Saldo atual: R$ {saldo:.2f}, Empréstimo disponível: R$ {emprestimo:.2f}")
        else:
            print("Operação falhou! O valor informado é inválido.")

    elif opcao == "2":
        valor = float(input("Informe o valor do saque: "))

        excedeu_saldo = valor > saldo
        excedeu_limite = valor > limite
        excedeu_saques = numero_saques >= LIMITE_SAQUES
        saque_possivel = saldo + emprestimo >= valor

        if excedeu_limite:
            print("Operação falhou! O valor do saque excede o limite.")

        elif excedeu_saques:
            print("Operação falhou! Número máximo de saques excedido.")

        elif valor > 0:
            if saque_possivel:
                if excedeu_saldo:
                    diferenca = valor - saldo
                    saldo = 0
                    emprestimo -= diferenca
                    extrato += f"Saque: R$ {valor:.2f} (R$ {diferenca:.2f} emprestados)\n"
                else:
                    saldo -= valor
                    extrato += f"Saque: R$ {valor:.2f}\n"

                numero_saques += 1
                print(f"Saque realizado com sucesso! Saldo atual: R$ {saldo:.2f}, Empréstimo disponível: R$ {emprestimo:.2f}")
            else:
                print("Operação falhou! Você não tem saldo suficiente e o valor do empréstimo disponível não cobre o saque.")

        else:
            print("Operação falhou! O valor informado é inválido.")

    elif opcao == "3":
        print("\n================ EXTRATO ================")
        print("Não foram realizadas movimentações." if not extrato else extrato)
        print(f"\nSaldo: R$ {saldo:.2f}")
        print(f"Empréstimo disponível: R$ {emprestimo:.2f}")
        print("==========================================")

    elif opcao == "4":
        break

    else:
        print("Operação inválida, por favor selecione novamente a operação desejada.")
```
